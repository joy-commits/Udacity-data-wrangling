## Report: act_report

WeRateDogs is a Twitter handle that is a legit verified source for professional dog ratings. Active since November 2015 and with total tweets and retweets of over 15,700, it has a followership of about 9.2 million followers as at June 2022. 
The WeRateDogs account also classifies dogs into four (4) stages. These stages are the pupper, puppo, doggo and floofer stages. The puppers are puppies. The puppos are the "older puppers". The doggos are the "older puppos". The floofers are the "fluffy" or "hairy" dogs.
The puppers are the most liked dogs based in tweets from the WeRateDogs handle. They also the most retweeted dogs.
